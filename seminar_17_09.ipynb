### Заливка

```python
# Простейшая заливка - заливается область между графиком функции и осью абсцисс
x = range(10)
y = [point**2 for point in x]
plt.fill(
    x,  # абсциссы точек
    y  # ординаты точек
)
plt.show()
```

```python
# Заливка между двумя графиками функций
x = range(10)
y1 = [point**2 for point in x]
y2 = [-point**2 for point in x]
plt.fill_between(
    x,  # абсциссы точек
    y1,  # ординаты первой кривой 
    y2  # ординаты второй кривой
)
plt.show()
```

```python
plt.plot(
    x,  
    y1,
)
plt.plot(
    x,  
    y2,
)
plt.show()
```

```python
x1 = range(10)
x2 = range(10, 20)
y = [point**2 for point in x]
plt.fill_betweenx(
    y,  # ОРДИНАТЫ точек
    x1,  # АБСЦИССЫ первой кривой 
    x2  # АБСЦИССЫ второй кривой
)
plt.show()
```

```python
plt.plot(
    x1,  
    y,
)
plt.plot(
    x2,  
    y,
)
plt.show()
```

### Сложные примеры

Разберем пример показывающий какие элементы рисунка могут быть на области рисования и какие у них могут быть настройки. Мы нарисуем рисунок на котором будет
 - одна область рисования
 - заголовок графика
 - график функции $\cos(x^2)$ и $\sin(x^2)$
 - аннотация (описание внутри графика, рассказывающее, что происходит на графике)
 - текст в рамке
 - подписи к осям X и Y
 - легенды к графикам
 - координатная сетка
 

```python
import numpy as np
```

```python
#  Генерируем данные
x = np.arange(0.0, 2*np.pi, 0.01)
y1 = np.cos([point**2 for point in x])
y2 = np.sin([point**2 for point in x])

# Создать рисунок 
fig = plt.figure(
    figsize=(16, 9),  # размером 16 на 9 дюймов
#     facecolor='whitesmoke',  # c подложкой цвета белый дым
)

# # Добавить заголовок графика
plt.title('Заголовок графика', fontsize=20)

# Нарисовать cos(x^2)
plt.plot(
    x, y1,
    color='blue',  # Цвет синий
    linestyle='--',  # Стиль - пунктирная линия
    alpha=0.4,  # Непрозрачность
    linewidth=5,  # Толщина
    label='Легенда про $\cos(x^2)$' # Подпись в легенде. Можно использовать TeX!
)

# Нарисовать sin(x^2)
plt.plot(
    x, y2,
    color='red',  # Цвет синий
    linestyle='--',  # Стиль - пунктирная линия
    alpha=0.4,  # Непрозрачность
    linewidth=5,  # Толщина
    label='Легенда про $\sin(x^2)$' # Подпись в легенде. Можно использовать TeX!
)

# Поставить легенду
plt.legend(
    loc='lower left',  # Снизу слева
    borderaxespad=5  # С отступом 5 от осей
)


# Добавим аннотацию
plt.annotate(
    'Аннотация для нашего графика',
    xy=(1.15, -0.25),  #  точка куда будет указывать стрелка
    arrowprops=dict(arrowstyle='->'),
    xytext=(0, -0.5)  # расположение аннотации
)

# Добавить подписи к осям X и Y
plt.xlabel(
    'Ось X',  # Текст
    fontdict=dict(family='serif', color='darkred', weight='normal', size=16)  # Настройки шрифта
)
plt.ylabel(
    'Ось  Y',
    fontdict=dict(family='monospace', color='peru', weight='light', size=25)
)

# # Добавить текст в квадрате
plt.text(
    2.0, 0.2,  # Позиция начала текста
    'Просто текст в квадрате',  # Собственно сам текст
    fontsize=26,  # Размер шрифта
    color='firebrick',  # Цвет букв
    rotation=10,
    bbox=dict(  # Параметры рамки ограничивающей текст
        alpha=0.4, # Прозрачность фона рамки
        edgecolor='slateblue',  # цвет рамки
        facecolor='azure',  # цвет подложки
        pad=1., # отступ рамки от текста
        boxstyle='round'  # форма рамки
    )
)

# # Добавить координатную сетку
plt.grid(True)

plt.show()
```

### Несколько областей рисования

```python
fig = plt.figure(figsize=(16, 9))
# fig.title('Заголовок графика', fontsize=20)

# Нарисуем основной график
x = np.linspace(0.001, 4, 400)
big_ax = fig.add_axes([0, 0, 1, 1])
big_ax.plot(
    x,
    np.sin(x**2),
    label='$sin(x^2)$'
)
big_ax.text(
    0, -1,
    'Большой график',
    fontsize=30
)

big_ax.legend(loc='upper right',)
big_ax.grid(True)

# Нарисуем вспомогательный график
small_ax = fig.add_axes([0.1, 0.2, 0.2, 0.3])
small_ax.plot(
    x,
    np.sin(x**2),
    color='green'
)
small_ax.text(
    0, -1,
    'Маленький график',
    fontsize=10
)
small_ax.grid(True)


plt.show()
```

### Несколько графиков на одном рисунке 

```python
from matplotlib import gridspec  # GridSpec определен там

# Создаем картинку
fig = plt.figure(figsize=(16, 9))
fig.suptitle('Zoo of hyperbolic functions', fontsize=20)

# Создаем спецификацию
spec = gridspec.GridSpec(nrows=3, ncols=4)
# В метод add_suplot можно передать slicing от спеки. 
# Тем самым создастся область рисования занимающая ячейки соответствующие slicing'у 
fig.add_subplot(spec[0, :3])  # Добавить область рисования занимающую первые три ячейки первой строки
fig.add_subplot(spec[:2, 3])
fig.add_subplot(spec[2, 1:])
fig.add_subplot(spec[1:, 0])
fig.add_subplot(spec[1, 1:3])

funcs = [np.sinh, np.tanh, np.cosh, np.arcsinh, np.arccosh]
points = np.linspace(-2, 2, 100)

for ax, func in zip(fig.axes, funcs):
    ax.plot(
        points,
        func(points),
        label='{}(x)'.format(func.__name__)
    )
    ax.legend(loc='best')
    ax.grid(True)

plt.show()
del fig, spec, ax, func
```
